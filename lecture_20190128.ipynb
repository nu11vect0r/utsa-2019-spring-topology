## The Real Numbers
$(\mathbb{R}, +, \cdot, <, 0, 1)$ is a mathematical structure.  $\mathbb{R}$ is the universe of the structure, $+$ and $\cdot$ are binary operations, $<$ is a relation, and 0, 1 are constants.

Fin a nontrivial ultrafilter $\U$ on $\mathbb{N}$.  Consider $\mathbb{R}^\mathbb{N}$, the set of all infinite sequences of real numbers.  We extend $\mathbb{R}$ to a larger field $\mathbb{R}^*$ as follows.  Let

$$
x = (x_1, x_2, \ldots) \\
y = (y_1, y_2, \ldots)
$$

be sequences in $\mathbb{R}^\mathbb{N}$.

**Definition**.  We say that $x$ and $y$ are *equal almost everywhere* if

$$
\{n | x_n = y_n\} \in \U.
$$

If $x$ and $y$ are equal almost everywhere, we write 

$$
x = y \textrm{ a.e.}
$$

**Remark** Being equal almost everywhere is an equivalence relation.

**Proof.** Reflexive.  $x = x \textrm{ a.e.}$ because $\{n | x_n = y_n\} = \mathbb{N} \in \U$.
Symmetric.  Trivial.  $\{n | x_n = y_n\} = \{n | y_n = x_n\}$.
Transitive.  Suppose $x = y \textrm{ a.e.}$ and $y = z \textrm{ a.e.}$.  Then 

$$
A = {n | x_n = y_n} \in \U \\
B = {n | y_n = z_n} \in \U
$$

Hence $A \cap B \in \U$ and $\{n | x_y - z_n\} \supseteq A \cap B$.  By the properties of filters, $\{n | x_n = z_n\} \in \U$.

**Observations**  Only used the fact that $\U$ is a filter (vs. an ultrafilter).  Also, didn't use any particular properties of $\mathbb{N}$ as the index set.

**Definition**  Define the non-standard reals (hyperreals) as the equivalences classes of

$$
\mathbb{R}^* = \mathbb{R}^\mathbb{N} \Big/ = (\textrm{a.e.}).
$$

Our goals
1. Prove that $\mathbb{R} \subseteq \mathbb{R}^*$.
1. Define $+^*$, $\cdot^*$, $0^*$, $1^*$, and $<^*$ such that the structure $(\mathbb{R}^*, +^*, \cdot^*, <^*, 0^*, 1^*)$ extends the structure $(\mathbb{R}, +, \cdot, <, 0, 1)$.

We embed $\mathbb{R}$ into $\mathbb{R}^*$ as follows.  Define

$$
f: \mathbb{R} \rightarrow \mathbb{R}^*
$$

by $x \to (x, x, \ldots)$, actually, the equivalence class of $(x, x, \ldots)$.  Note that $f$ is one-to-one because

$$
x \neq y \Rightarrow (x, x, \ldots) \neq (y, y, \ldots) \textrm{ a.e.}
$$

* Let x, y \in \mathbb{R}^*$ with

$$
x = (x_1, x_2, \ldots) \\
y = (y_1, y_2, \ldots)
$$

and $x \leq^* y$ if and only if $\{n | x_n \leq y_n \} \in \U$.  (Show that this is well-defined.)

* Similarly, we extend $+$ to $+^*$ by defining

$$
x +^* y = z \textrm{ iff } \{n | x_n + y_n = z_n \} \in \U.
$$

A routine argument shows this definition is well-defined.  Also, multiplication, $\cdot^*$ is definied similarly. 

For any $r \in \mathbb{R}$, we let $r^*$ be the equivalence class of $(r, r, \ldots)$.  It is routine to prove that $(\mathbb{R}^*, +^*, \cdot^*, <^*, 0^*, 1^*)$ is an ordered field.

**Remark**  $\mathbb{R}^*$ contains both infinite and infinitesimal elements.  Consider $x = (1, 2, 3, \ldots)$.  Note that

$$
x > 1 \textrm{ a.e.} \\
x > 2 \textrm{ a.e.} \\
etc.
$$
 
Thus "$x$ is infinite".  Similarly, let $\frac{1}{x} = (1, \frac{1}{2}, \frac{1}{3}, \ldots)$.  Note that

$$
\frac{1}{x} < 1 \textrm{ a.e.} \\
\frac{1}{x} < \frac{1}{2} \textrm{ a.e.} \\
etc.
$$
