## Filters

### Finite Intersection Property

__Definition__ Let $I$ be a set, let $\mathcal{C}$ be a collection of subsets of $I$.  We will say that $\mathcal{C}$ has the *finite intersection property* (or that $\mathcal{C}$ is a *consistent*) if whenever $A_1, \ldots, A_n \in \mathcal{C}$, we have

$$
\bigcap_{i=1}^n A_i \neq \emptyset.
$$

__Theorem__ If $I$ is a set and $\mathcal{C} \subseteq \mathcal{P}(I)$ has the finite intersection property, then there is a filter $\F$ on $I$ such that $\F \supseteq \mathcal{C}$. 

Note that the converse of the theorem is trivial.  If $\mathcal{C}$ is contained in a filter, then $\mathcal{C}$ satisfies the finite intersection property.

*Proof.*  Assume $\mathcal{C} \subseteq \mathcal{P}(I)$ has the finite intersection property.  Define

$$
\F = \{A \subseteq I | A \textrm{ contains a finite intersection of elements in } \mathcal{C} \}
$$

then it is easy to check that $\F$ is a filter. $\qedsymbol$

### Tarski's Theorem

Recall that $\U$ is an ultrafilter on $I$ if for every subset $A \subseteq I$, we have either

$$
A \in \U \textrm{ or } A^c \in \U.
$$

__Theorem__ (Tarski) For every filter $\F$ on $I$, there exists an ultrafilter $\U$ on $I$ such that $\U \supseteq \F$.

In order to prove this theorem, we need the following lemma.

__Lemma__ Let $\F$ be a filter on $I$, and let $A \subseteq I$ then either $\F \cup \{A\}$ or $\F \cup \{A^c\}$ has the finite intersection property.

*Proof.* Assume $\F \cup \{A\}$ doesn't have the finite intersection property.  Fix $F_1, \ldots, F_n \in \F$ such that 

$$
\left(\bigcap F_i\right) \cap A = \emptyset.
$$

Then $\bigcap F_i \subseteq A^c$ and their intersection is in $\F$. $\qedsymbol$

*Proof of Tarski's Theorem.* Let $A_0, A_1, \ldots, A_i, \ldots$ where $i < \alpha$ ($\alpha$ is some ordinal) be a well-ordered list of all subsets of $I$.  Applying the lemma inductively, we obtain a family of filters,

$$
\F \subseteq \F_0 \subseteq \F_1 \subseteq \cdots \subseteq \F_i \cdots \textrm{ for } (i < \alpha)
$$

such that for each $i < \alpha$ either $A_i$ or $A_i^c$ is contained in $\F_i$.  Let

$$
\U = \bigcup_{i < \alpha} \F_i.
$$

We claim $\U$ is an ultrafilter.  Clearly, $I \in \F$ so $I \in \U$.  Also, $\U$ is closed under supersets because each $\F_i$ is closed under supersets.  Finally, $\U$ is closed under finite intersections, because if $A, B \in \U$, there are $i, j < \alpha$ such that $A \in \F_i$ and $B \in \F_j$, so

$$
A, B \in \F_{\max(i, j)}.
$$

Since $\F_{\max(i, j)}$ is a filter it contains $A \cap B \subseteq \U$.  $\qedsymbol$

__Proposition__ Let $I$ be a set and let $\F$ be a filter on $I$, then the following are equivalent.

1. $\F$ is an ultrafilter.
1. $\F$ is a maximal filter on $I$, i.e., there is no filter $\F'$ on $I$ such that $\F \subsetneq \F'$.

*Proof $(1 \Rightarrow 2)$.*  Prove the contrapositive, suppose not (2), then fix a filter $\F' \supsetneq \F$.  Fix $A \in \F' \backslash \F$.  Now $A \not\in \F$ and $A^c \not\in \F$.  Otherwise, $A^c \in \F'$ which can't occur since $A \in \F'$.  Hence, $\F$ is not an ultrafilter.

*$(2 \Rightarrow 1)$* Suppose (1) doesn't hold.  Fix $A \subseteq I$ such that $A \not\in \F$ and $A^c \not\in \F'$.  By the lemma, either 

$$
\F \cup \{A\} \textrm{ or } \F \cup \{A^c\}
$$

is a filter, so $\F$ is not maximal.

### Zorn's Lemma

__Lemma (Zorn)__  Let $(\Sigma, \leq)$ be a nonempty, partially ordered set such that every linearly ordered subset of $\Sigma$ has an upper bound.  Then $(\Sigma, \leq)$ has a maximal element.

*Proof of Tarski's Theorem using Zorn's Lemma.*  Fix a filter $\F$ on $I$.  Let

$$
\Sigma = \{ \G | \G \textrm{ is a filter on } I \textrm{ and } \G \supseteq \F \}
$$

$\Sigma$ is nonempty (since it contains $\F$).  Order $\Sigma$ by inclusion, i.e., for $\G, \G' \in \Sigma$

$$
\G \leq \G' \textrm{ iff } \G \subseteq \G'.
$$

Every linearly ordered subset $\Sigma_0 \subseteq \Sigma$ has an upper bound, namely

$$
\bigcup_{\G \in \Sigma_0} \G.
$$

Hence, by Zorn's Lemma, $\Sigma$ has a maximal element.  By the preceeding proposition, such a maximal element is an ultrafilter.
